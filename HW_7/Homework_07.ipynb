```python
import time
from tensorflow.keras.datasets import mnist, fashion_mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255

(X_fashion_train, y_fashion_train), (X_fashion_test, y_fashion_test) = fashion_mnist.load_data()

X_fashion_train = X_fashion_train.astype('float32') / 255
X_fashion_test = X_fashion_test.astype('float32') / 255

from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

y_fashion_train = to_categorical(y_fashion_train)
y_fashion_test = to_categorical(y_fashion_test)

from tensorflow.keras.models import Sequential

cnn_mist = Sequential()
cnn_fashion = Sequential()

from tensorflow.keras.layers import Conv2D, Dense, Flatten, MaxPooling2D

cnn_mist.add(Conv2D(filters=64, kernel_size= (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn_mist.add(MaxPooling2D(pool_size=(2, 2)))
cnn_mist.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
cnn_mist.add(MaxPooling2D(pool_size=(2, 2)))
cnn_mist.add(Flatten())
cnn_mist.add(Dense(units=128, activation='relu'))
cnn_mist.add(Dense(units=10, activation='softmax'))
cnn_mist.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time_mnist = time.time()

cnn_mist.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.1)

training_time_mnist = time.time() - start_time_mnist

cnn_fashion.add(Conv2D(filters=64, kernel_size= (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn_fashion.add(MaxPooling2D(pool_size=(2, 2)))
cnn_fashion.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
cnn_fashion.add(MaxPooling2D(pool_size=(2, 2)))
cnn_fashion.add(Flatten())
cnn_fashion.add(Dense(units=128, activation='relu'))
cnn_fashion.add(Dense(units=10, activation='softmax'))
cnn_fashion.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time_fashion = time.time()

cnn_fashion.fit(X_fashion_test, y_fashion_train, epochs=5, batch_size=64, validation_split=0.1)

training_time_fashion = time.time() - start_time_fashion

loss_mist, accuracy_mist = cnn_mist.evaluate(X_test, y_test)
loss_fashion, accuracy_fashion = cnn_fashion.evaluate(X_fashion_test, y_fashion_test)

print(f'MNIST Accuracy: {accuracy_mist}\nMNIST Training Time: {training_time_mnist}')
print(f'Fashion MNIST Accuracy: {accuracy_fashion}\nFashion MNIST Training Time: {training_time_fashion}')

(X_train_remove, y_train_remove), (X_test_remove, y_test_remove) = mnist.load_data()
(X_train_add_4096, y_train_add_4096), (X_test_add_4096, y_test_add_4096) = mnist.load_data()

X_test_remove = X_test_remove.astype('float32') / 255
X_train_remove = X_train_remove.astype('float32') / 255

X_test_add_4096= X_test_add_4096.astype('float32') / 255
X_train_add = X_train_add_4096.astype('float32') / 255

y_train_remove = to_categorical(y_train_remove)
y_test_remove = to_categorical(y_test_remove)

y_train_add_4096 = to_categorical(y_train_add_4096)
y_test_add_4096 = to_categorical(y_test_add_4096)



cnn_remove = Sequential()

cnn_remove.add(Conv2D(filters=64, kernel_size= (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn_remove.add(MaxPooling2D(pool_size=(2, 2)))
cnn_remove.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
cnn_remove.add(MaxPooling2D(pool_size=(2, 2)))
cnn_remove.add(Flatten())
cnn_remove.add(Dense(units=10, activation='softmax'))
cnn_remove.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time_remove = time.time()

cnn_remove.fit(X_train_remove, y_train_remove, epochs=5, batch_size=64, validation_split=0.1)

training_time_remove = time.time() - start_time_remove

cnn_add = Sequential()

cnn_add.add(Conv2D(filters=64, kernel_size= (3, 3), activation='relu', input_shape=(28, 28, 1)))
cnn_add.add(MaxPooling2D(pool_size=(2, 2)))
cnn_add.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
cnn_add.add(MaxPooling2D(pool_size=(2, 2)))
cnn_add.add(Flatten())
cnn_add.add(Dense(units=128, activation='relu'))
cnn_add.add(Dense(units=4096, activation='relu'))
cnn_add.add(Dense(units=10, activation='softmax'))
cnn_add.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time_add = time.time()

cnn_add.fit(X_train_add_4096, y_train_add_4096, epochs=5, batch_size=64, validation_split=0.1)

training_time_add_4096 = time.time() - start_time_add

loss_remove_mist, accuracy_remove_mist = cnn_remove.evaluate(X_test_remove, y_test_remove)
loss_add_mist, accuracy_add_mist = cnn_add.evaluate(X_test_add_4096, y_test_add_4096)

print(f'Layer Removed MNIST Accuracy: {accuracy_remove_mist}\nLayer Removed MNIST Training Time: {training_time_remove}')
print(f'Added MNIST Accuracy: {accuracy_add_mist}\nAdded MNIST Training Time: {training_time_add_4096}')
```
